```python
# Import necessary libraries
import pandas as pd
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import numpy as np
from sklearn.inspection import DecisionBoundaryDisplay
```

```python
# Read penguins data from CSV file
dataset = pd.read_csv("data/penguins_dataset.csv")
dataset.head(5)
```

```python
# Select the features and prediction columns, ensure 'species' is categorical
X = dataset[["culmenLength", "culmenDepth"]]
y = pd.factorize(dataset["species"])[0] # Convert species to numerical values 
```

```python
# Plot the data
plt.figure(figsize=(6,4))
plt.scatter(X['culmenLength'], X['culmenDepth'], c=y, cmap='viridis')
plt.title('Penguins Data: Culmen Lenth vs Culmen Depth')
plt.xlabel('Culmen Length (mm)')
plt.ylabel('Culmen Dept (mm')
plt.show()
```

```python
# Create and train the Logistic Regression model on all data
model = LogisticRegression(max_iter=10000)
model.fit(X,y)
```

```python
# Plot the data with the computed decision boundary
disp= DecisionBoundaryDisplay.from_estimator(
    model, X, cmap="viridis", alpha=0.3, response_method="predict"
)
plt.scatter(X["culmenLength"], X["culmenDepth"], c=y, cmap="viridis")

plt.xlabel("Culmen Length")
plt.ylabel("Culmen Depth")
plt.title("Logistic Regression Decision Boundary")
plt.show()
```
