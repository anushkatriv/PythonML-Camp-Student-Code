```python
class FruitJersey: 
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display(self):
        print(f"Item: {self.name}, Price: ${self.price:.2f}")


grape = FruitJersey("grapes", 6)
watermelon = FruitJersey("watermelons", 10)
blueberry = FruitJersey("blueberries", 67)


grape.display()
watermelon.display()
blueberry.display()
```

```python
import random 
class Calculator:
    def __init__(self, model, function, num_buttons):
        self.model = model
        self.function = function
        self.num_buttons = num_buttons
    
    def print_info(self):
        print("the model is",self.model)
        print("the function is", self.function)
        print("the num_button is", self.num_buttons)
    
    def sum (self, a, b): 
        print(a/b)

    def subtract (self, a, b): 
        if a-b<0: 
            print("it's negative")
        elif a-b>0: 
            print("it's positive") 
        else:
            print("it's zero")
        print(a-b)

    def multiply (self, a, b):
        print(a*b)

    def divide (self, a, b):
        print(a/b)
    def remainder (self, a, b): 
        print(a%b)
    def random (self, lower, upper, num):
        if lower>upper: 
            print("Upper bound should be larger than lower bound")
            return None
        else:
            x=0
            rand_nums=[]
            while x<num:
                temp=(random.randrange(lower, upper))
                print(temp)
                rand_nums.append(temp)
                x+=1
            print(rand_nums[0])
            print(rand_nums[-1])
            return rand_nums
            
        



            

calc= Calculator("inspire_cx", True, 50)
calc.print_info()
calc.sum(37, 93)
calc.subtract(84, 22)

product = calc.multiply(23, 54)

quotient = calc.divide(15,3)
remainder = calc.remainder(22,7)
calc.random(10,50,7)
```
